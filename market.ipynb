```python
import time

driver = webdriver.Chrome()
driver.get('https://www.tiendeo.de/Filialen/brandenburg-an-der-havel/supermarkt')
time.sleep(3)
# mehr geschaefte button
for i in range(3):
    driver.find_element(By.CSS_SELECTOR, '[class = "sc-eGugkK WDlhU"]').click()
    time.sleep(5)
```

```python
element = driver.find_elements(By.CSS_SELECTOR, 'li[data-testid="itemmainStores"]')

market_list = []

for i in element:
    
    dic = {}
    dic["title"] = i.find_element(By.CSS_SELECTOR,'li[data-testid="itemmainStores"] h3').text # Supermarkt titel
    dic["address"] = i.find_element(By.CSS_SELECTOR,'li[data-testid="itemmainStores"] span').text # adresse
    dic["logo"] = i.find_element(By.CSS_SELECTOR, 'li[data-testid="itemmainStores"] img').get_attribute('src') # logo url

    market_list.append(dic)


df = pd.json_normalize(market_list)
df


```
